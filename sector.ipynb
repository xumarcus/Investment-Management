```python
df = pd.read_csv('semiconductors.csv')
df
```

```python
returns_3mo = ((df.values[-1][1:] - df.values[0][1:]) / df.values[0][1:]).tolist()
returns_3mo
```

```python
weights_3mo = (df.values[0][1:] / np.sum(df.values[0][1:])).tolist()
weights_3mo
```

```python
weighted_returns_3mo = np.multiply(returns_3mo, weights_3mo)
print(f"Weighted Returns of 3 months investment in semiconductors are {'%.2f' % (np.sum(weighted_returns_3mo) * 100)}%")
```

```python
df = df.set_index('date')
df    
```

```python
log_returns = np.log(df / df.shift(1)).dropna()
log_returns
```

```python
cov_matrix = log_returns.cov() * df.shape[0]
cov_matrix
```

```python
weights_3mo = np.array(weights_3mo)
portfolio_variance = weights_3mo.T @ cov_matrix @ weights_3mo
print(f"Standard Deviation: {np.sqrt(portfolio_variance) * 100}%")
```

```python
rf = yq.Ticker('^TNX').price['^TNX']['regularMarketPrice'] / 100
rf
```

```python
sharpe_ratio = (np.sum(weighted_returns_3mo) - rf) / np.sqrt(portfolio_variance)
sharpe_ratio
```

```python
def MC_VaR(amount, returns, vol, period, iterations):
    
    end = amount * np.exp((returns - .5 * vol ** 2) * period + 
                     vol * np.sqrt(period) * np.random.standard_normal(iterations))
    ending_values = end - amount
    return ending_values
```

```python
np.sum(df.values[0][1:])
```

```python
iterations = 50000
var = MC_VaR(np.sum(df.values[0][1:]), np.sum(weighted_returns_3mo), np.sqrt(portfolio_variance), 1 / df.shape[0], iterations)
var
```

```python
percentiles = [1,5,10]
v99,v95,v90 = np.percentile(var, percentiles) * -1
print(f"At 99% confidence level, loss will not exceed {v99:,.2f}")
print(f"At 95% confidence level, loss will not exceed {v95:,.2f}")
print(f"At 90% confidence level, loss will not exceed {v90:,.2f}")
```

```python
percentiles = [1,5,10]
ed99 = np.average([x for x in var if x < v99 * -1])
ed95 = np.average([x for x in var if x < v95 * -1])
ed90 = np.average([x for x in var if x < v90 * -1])
print(f"At 99% confidence level, expected downfall averages at {ed99:,.2f}")
print(f"At 95% confidence level, expected downfall averages at {ed95:,.2f}")
print(f"At 90% confidence level, expected downfall averages at {ed90:,.2f}")
```
